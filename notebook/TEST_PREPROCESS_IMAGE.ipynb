```python
from __future__ import print_function, division
from torch.utils.data import Dataset
from skimage import io
import glob
import os
import os.path as osp
import torch
from torchvision import transforms
import cv2
import pandas as pd 
import numpy as np
import random
from PIL import Image
import albumentations as A
import glob
import matplotlib.pyplot as plt

WORKING_DIR = '/home/ntphat/projects/medico/medico'
os.chdir(WORKING_DIR)

DATA_DIR = './data'
TRAIN_IMG = osp.join(DATA_DIR, 'test_images')
```

```python
list_img_paths = glob.glob(TRAIN_IMG +'/*')
sample_img_path = random.choice(list_img_paths)
sample_img = Image.open(sample_img_path)
print(sample_img.size) #W, H
display(sample_img)
```

```python
new_img = cv2.bilateralFilter(np.array(sample_img), 9, 75, 75)
new_img = Image.fromarray(new_img).convert("RGB")
display(new_img)
```

```python
# dilation
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(np.array(sample_img),kernel,iterations = 1)
dilation = Image.fromarray(dilation).convert("RGB")
display(dilation)
```

```python
gradient = cv2.morphologyEx(np.array(sample_img), cv2.MORPH_GRADIENT, kernel)
gradient = Image.fromarray(gradient).convert("RGB")
display(gradient)
```
